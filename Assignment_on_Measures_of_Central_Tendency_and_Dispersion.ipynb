**3. Who scores the most on average for math, reading and writing based on**

**Gender**

```python
data1=stperform.groupby('gender')
```

```python
data1[['math score','reading score','writing score']].mean()
```

The average math score of males is more than that of females. In the case of reading score the average score of females is more than that of males. For writing score the average score of females is more than that of males.  

**Test preparation course**

```python
data2=stperform.groupby('test preparation course')
```

```python
data2[['math score','reading score','writing score']].mean()
```

The average 'math score','reading score' and 'writing score' of people completed test preparation course is more than that of those who didn't complete it.
