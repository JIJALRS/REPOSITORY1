**● The company wishes to clarify whether there is any increase in sales after
stepping into digital marketing.**

Let m1 be the average sales after stepping into digital marketing and m2 be the average sales before stepping into digital marketing.we need to clarify whether there is any increase in sales after stepping into digital marketing.
Therefore our hypothesis is

H0: m1 <= m2

vs

H1: m1>m2

since sample size of both the groups is less than 30 we can use independent two sample T-test for testing the hypothesis

```python
#Importing the function for T test
from scipy.stats import ttest_ind
```

```python
t_stat,p_value=ttest_ind(sa['Sales_After_digital_add(in $)'],sa['Sales_before_digital_add(in $)'],alternative="greater")
```

```python
t_stat
```

```python
p_value
```

```python
if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

Hence we reject the null hypothesis at 5% level of significance. Therefore from the given sample observations we can conclude that there is a statistically significant increase in sales after stepping into digital marketing.

**● The company needs to check whether there is any dependency between
the features “Region” and “Manager”.**

Here we need to use the Chisquare test of Independence to test whether there is any dependency between the features “Region” and “Manager”. The hypothesis to be tested is

H0: Region and Sales are independent of each other

vs

H1: Region and Sales are dependent on each other

```python
#Since the data is not in the required format i need to reshape the data using melt function
sa1=pd.melt(sa,id_vars=['Manager','Region'],value_vars=['Sales_before_digital_add(in $)','Sales_After_digital_add(in $)'],value_name='Sales')
```

```python
sa1
```

```python
#Now use the crosstab function to create a crosstab between Manager and Region taking sales as the frequency
ctab=pd.crosstab(sa1['Region'],sa1['Manager'],values=sa1['Sales'],aggfunc='sum')
```

```python
ctab
```

```python
#Importing the function for chisquare test of independence
from scipy.stats import chi2_contingency
```

```python
chistat,pvalue,dof,expected=chi2_contingency(ctab)
```

```python
chistat
```

```python
pvalue
```

```python
dof
```

```python
if pvalue < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

Therefore the variables manager and region are dependent statistically at 4 degrees of freedom and 5% level of significance
